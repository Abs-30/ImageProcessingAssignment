**Identitiy Transformation**

g(x) = F(x)

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

im = cv.imread('images/jeniffer.jpg', cv.IMREAD_GRAYSCALE)
assert im is not None

#image transform formula
t = np.arange(0,256).astype('uint8')        #mapping each 256 pints to 256 
fig, ax = plt.subplots()
ax.plot(t)
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel(r'Output, $\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
plt.show()

#showing image
cv.namedWindow('Image', cv.WINDOW_AUTOSIZE)
cv.imshow('Image', im)
cv.waitKey(0)
image_transform = cv.LUT(im, t)
cv.imshow('Image', image_transform)
cv.waitKey(0)
cv.destroyAllWindows()

# no transformations occur because the transform and the original image are the same thing

```

**Negative Transformation**
g(x)=255 - f(x)

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

im = cv.imread('images/jeniffer.jpg', cv.IMREAD_GRAYSCALE)
assert im is not None

#image transform formula
t = np.arange(255, -1, -1).astype('uint8')        #start at 255, going until 0, => in steps of zero
fig, ax = plt.subplots()
ax.plot(t)
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel(r'Output, $\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
plt.show()

#showing image
cv.namedWindow('Image', cv.WINDOW_NORMAL)
cv.imshow('Image', im)
cv.waitKey(0)
image_transform = cv.LUT(im, t)
cv.imshow('Image', image_transform)
cv.waitKey(0)
cv.destroyAllWindows()
```

**Intensity Windowing**

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

im = cv.imread('images/jeniffer.jpg', cv.IMREAD_GRAYSCALE)
assert im is not None

c = np.array([(100,50) , (150,200)])
#Creating the regions
t1 = np.linspace(0, c[0,1],c[0,0] +1 -0).astype('uint8') 
print(len(t1))       #start at 255, going until 0, => in steps of zero
t2 = np.linspace(c[0,1]+1, c[1,1],c[1,0] - c[0,0]).astype('uint8') 
print(len(t2))
t3 = np.linspace(c[1,1] +1, 255,255 - c[1,0]).astype('uint8') 
print(len(t3))

#transform
transform = np.concatenate((t1,t2), axis=0).astype('uint8')
transform = np.concatenate((transform, t3), axis=0).astype('uint8')
print(len(transform))

#fig
fig, ax = plt.subplots()
ax.plot(transform)
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel(r'Output, $\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
plt.show()

#showing image
cv.namedWindow('Image', cv.WINDOW_NORMAL)
cv.imshow('Image', im)
cv.waitKey(0)
image_transform = cv.LUT(im, transform)
cv.imshow('Image', image_transform)
cv.waitKey(0)
cv.destroyAllWindows()
```
