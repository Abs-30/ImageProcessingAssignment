```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

im = cv.imread('images/jeniffer.jpg', cv.IMREAD_GRAYSCALE)
assert im is not None

#image transform formula
t = np.arange(0,256).astype('uint8')        #mapping each 256 pints to 256 
fig, ax = plt.subplots()
ax.plot(t)
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel(r'Output, $\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
plt.show()

#showing image
cv.namedWindow('Image', cv.WINDOW_AUTOSIZE)
cv.imshow('Image', im)
cv.waitKey(0)
image_transform = cv.LUT(im, t)
cv.imshow('Image', image_transform)
cv.waitKey(0)
cv.destroyAllWindows()

# no transformations occur because the transform and the original image are the same thing
```
